```python
import matplotlib.pyplot as plt
import seaborn as sns
import json
import ast
import numpy as np
from collections import defaultdict

# Load and parse the data from file
with open("./report_modified_cacdc.txt", "r") as f:
    text = f.read().split("\n")
dict_arr = [ast.literal_eval(strv) for strv in text if strv]

with open("./report.txt", "r") as f:
    text = f.read().split("\n")
dict_arr2 = [ast.literal_eval(strv) for strv in text if strv]
```

```python
forg = [dictv["forg_val"] for dictv in dict_arr]
util = [dictv["util_val"] for dictv in dict_arr]
k_frac = [dictv["k_frac"] for dictv in dict_arr]
forg2 = [dictv["Forg_Accuracy"] for dictv in dict_arr2]

sns.lineplot(x=k_frac, y=forg, label='$Acc_{aff} \ random$', marker='o')
sns.lineplot(x=k_frac, y=forg2, label='$Acc_{aff} \ original$', marker='o')
sns.lineplot(x=k_frac, y=util, label='$Acc_{rem}$', marker='o')
plt.ylim(0, 1)
plt.show()
```
