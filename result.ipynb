```python
def getarr(filename):
    with open(filename, "r") as f:
        arr=f.read().split("\n")
    text=[]
    for item in arr:
        try:
            text.append(float(item))
        except:
            continue
    return text
```

```python
import matplotlib.pyplot as plt
# cacdc= getarr("./cacdc.txt")
scrub= getarr("./scrub2.txt")
cacdc_a= getarr("./ascent.txt")
cacdc_d= getarr("./descent.txt")
# cacdc= getarr("./results3.txt")

# plt.plot(yaum, label="yaum")
plt.plot(scrub, label="scrub")
plt.plot(cacdc_a, label="ascent")
plt.plot(cacdc_d, label="descent")
plt.legend()
```
